```python
"""
This script will take a directory with .wav files and create new directories
with the name of each file + chunks appended to it that contains each .wav file
split into 1 second chunks. When using this script modify the "path" variable in
cell 3 to the location of the .wav data relative to the directory this script is
is running in.
"""
```

```python
import os
from pydub import AudioSegment
from pydub.utils import make_chunks
```

```python
# This is the path relative to the directory this script is running in.
path = "../../data/"
chunk_length_ms = 1000 # 1 second

with os.scandir(path) as dir:
    # Iterate over every ".wav" file located in path variable.
    for file in dir:
        if file.name.endswith(".wav") and file.is_file():
            audio = AudioSegment.from_file(file=file.path, format="wav")
            
            # Make chunks of one second for current .wav file.
            chunks = make_chunks(audio, chunk_length_ms)
            
            new_dir_name = f"{path}{file.name[:-4]}chunks"
            print(f"Current audio file: {file.name}")
            print(f"New dir name: {new_dir_name}")
            
            # Create new directory if it doesn't already exist.
            if not os.path.isdir(new_dir_name):
                os.makedirs(new_dir_name)
                pass
            
            # Save and export one second chunks to new directory.
            for i, chunk in enumerate(chunks):
                    chunk_name = f"{file.name[:-4]}chunk{i:05}.wav"
                    whole_path = f"{new_dir_name}/{chunk_name}"
                    print(f"Exporting {chunk_name}")
                    chunk.export(whole_path, format="wav")
```
